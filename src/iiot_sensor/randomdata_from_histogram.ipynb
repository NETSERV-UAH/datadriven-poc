```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
data=pd.read_csv('IIOT-temp-warn-max.csv')
```

```python
df=pd.DataFrame(data,columns=['room_id/id', 'noted_date', 'temp', 'out/in','Month','name_months','temp_warn'])
```

```python
df.describe()
```

```python
# Filtrar temperaturas por interior y exterior
temp_exterior = df[df['out/in'] == 'Out']['temp']
temp_interior = df[df['out/in'] == 'In']['temp']

# Crear histograma
plt.hist(temp_exterior, bins=10, alpha=0.5, label='Exterior')
plt.hist(temp_interior, bins=10, alpha=0.5, label='Interior')

# Añadir etiquetas y título
plt.xlabel('Temp')
plt.ylabel('Frec')
plt.title('Indoor and Outdoor Temperature Histogram')
plt.legend()
plt.minorticks_on()
plt.grid()

# Mostrar el histograma
plt.show()
```

```python
# Generar temperaturas aleatorias para el interior
random_interior = np.random.choice(temp_interior, size=100000, replace=True)

# Generar temperaturas aleatorias para el exterior
random_exterior = np.random.choice(temp_exterior, size=100000, replace=True)
```

```python
# Crear histograma aleatorio
plt.hist(random_exterior, bins=10, alpha=0.5, label='Exterior')
plt.hist(random_interior, bins=10, alpha=0.5, label='Interior')

# Añadir etiquetas y título
plt.xlabel('Temp')
plt.ylabel('Frec')
plt.title('Indoor and Outdoor Temperature random Histogram')
plt.legend()
plt.minorticks_on()
plt.grid()

# Mostrar el histograma
plt.show()
```

```python
# Guardamos las variables 
with open('temperaturas.pkl', 'wb') as f:
    pickle.dump((temp_exterior, temp_interior), f)
```
